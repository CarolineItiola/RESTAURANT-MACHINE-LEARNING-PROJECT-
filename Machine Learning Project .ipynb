```python
x = df.drop(['Aggregate rating', 'Rating Category'],axis=1)
y =df['Rating Category']
```

```python
# Write a function to convert aggregate rating to binary category
def encoder (rating):
    if rating > 4:
        return 1   # High rating
    else:
        return 0   # Low rating
```

```python
#The above rating category exhibit a class imbalance- 8430 low rated restaurants versus 1112 high rated restaurants
# So resample them
```

```python
#splitting dataset
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size=0.2, random_state=42)
```

```python
smote = SMOTE(random_state=42)
x_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)
```

```python
y_train_resampled.shape
```

```python
y_pred_lr = lr.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred_lr)*100)
print(precision_score(y_test,y_pred_lr)*100)
print(recall_score(y_test,y_pred_lr)*100)
print(f1_score(y_test,y_pred_lr)*100)
```

```python
#The model classified 86.49% of all restaurants correctly.Although the dataset initially exhibited class imbalance, applying SMOTE helped rebalance the 
#training data and improve detection of high rated restaurants.So, accuracy alone does not fully refelct the classifier's performance on minority class.
#Of all restaurants predicted as high rated, 44.67% were actually high rated.While the model is now identifying more high-rated restaurants, it also 
#mislabel a notable number of low rated restaurants as high rated, reflecting a decrease in precision
#The model correctly identified 81.48% of all actual high rated restautants. This is a substantial improvement comapred to the non-SMOTE model. This 
#demonstrates that the classifier is now much more effective in detecting high rated restaurants. 
#The F1 score balances precision and recall. A score of 57.70%% indicate improved overall performance, comparede to the non-SMOTE model and ths reflect
#a better trade-off between identifying high rated restaurants and controlling false positives. 
```
